Rosalind 10 ~ Compute the Number of Times a Pattern Appears in a Text

```python
Text = input("Text")
Pattern = input("Pattern")
count = 0
for i in range (len(Text) - len(Pattern)+1):
    if Text[i: i+len(Pattern)] == Pattern:
      count = count + 1
print(count)
```

Rosalind 11 ~ Find the Most Frequent Words in a String

```python
String = input ("string")
k = int(input("k"))
kmer_count = {}

for i in range (len(String)- k+1):
  kmer = String[i:i+k]
  if kmer in kmer_count:
    kmer_count[kmer] = kmer_count[kmer] + 1
  else:
    kmer_count[kmer] = 1

max(kmer_count.values())
for kmer, count in kmer_count.items():
  if count == max(kmer_count.values()):
    print(kmer)
```

Rosalind 12 ~ Find Patterns Forming Clumps in a String

```python
String = input('String')
k = int(input("k"))
L = int(input("L"))
t = int(input("t"))
clump_count = set()

for i in range (len(String)- L+1):
  clump = String[i:i+L]
  kmer_count = {}
  for j in range (len(clump)-k+1):
    kmer = clump[j:j+k]
    if kmer in kmer_count:
      kmer_count[kmer] = kmer_count[kmer] + 1
    else:
      kmer_count[kmer] = 1

  for kmer, count in kmer_count.items():
    if count >= t:
      clump_count.add(kmer)

for kmer in clump_count:
  print(kmer)
```

Rosalind 13 ~ Find a Position in a Genome Minimizing the Skew

```python
genome = input("genome")
skew = 0
min_skew = 0
min_loc = []

for i in range (len(genome)):
  if genome[i] == 'C':
    skew -= 1
  elif genome[i] == 'G':
    skew += 1
  if skew < min_skew:
        min_skew = skew
        min_loc = [i+1]
  elif skew == min_skew:
          min_loc.append(i+1)

print(min_loc)
```

Rosalind 14 ~ Compute the Hamming Distance Between Two Strings

```python
p = input("p")
q = input("q")
hamming = 0

for i in range (len(p)):
    if p[i] != q[i]:
      hamming = hamming + 1
print(hamming)
```
